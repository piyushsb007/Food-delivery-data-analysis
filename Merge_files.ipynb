```python
import pandas as pd
import json
```

```python
# Load Orders CSV
orders = pd.read_csv("orders.csv")
print("Orders Loaded:", orders.shape)
```

```python
# Load Users JSON
users = pd.read_json("users.json")
print("Users Loaded:", users.shape)
```

```python
import sqlite3
# Create, Connect db
conn = sqlite3.connect("reastaurant.db")
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS restaurants;")
conn.commit()

# load sql script
with open("restaurants.sql","r") as file:
    script = file.read()
# Execute Sql script
cursor.executescript(script)
conn.commit()
#Read restaurants table into pandas 
restaurants = pd.read_sql("SELECT * FROM restaurants",conn)
conn.close()
print("Restaurants loaded:", restaurants.shape)
```

```python
# Merge the data
merged = pd.merge(orders,users,on="user_id")
#file dataset 
fd = pd.merge(merged , restaurants , on="restaurant_id")
```

```python
# Save Final dataset
fd.to_csv("final_food_delivery_dataset.csv",index="False")
print(" final_food_delivery_dataset.csv created successfully!")
print("Shape:", fd.shape)
```
